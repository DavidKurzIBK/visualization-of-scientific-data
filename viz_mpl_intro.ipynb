# How to set up axes and plot

```python
%matplotlib inline

import numpy as np
import matplotlib.pyplot as plt
import pandas as pd
```

```python
# define some dummy data
df = pd.DataFrame({'x':range(5),'y':[1,3,4,2,4]})
df.head()
```

```python
# define plot params, linewidth (lw), markersize (ms) and color (c)
lw = 1           
ms = 7
c = (.1,.4,.95)
```

```python
# different ways of plotting
# 12 or 15 is the width of the whole notebook; 

plt.plot(df.x,df.y,'.-',c=c,ms=ms,lw=lw)
plt.grid()

# dataframe's plot method
df.plot(x='x',y='y',style='.-',c=c,ms=ms,lw=lw,grid=True)#,figsize=(7,5)

#rather define figure and axes explicitly 
fig = plt.figure(figsize=(7,7))
ax = fig.add_axes([0,0,1,.5])
# or in one line 
# ax = plt.figure(figsize=(7,7)).add_axes([0,0,1,.5])
ax.plot(df['x'],df['y'],'.-',c=c,ms=ms,lw=lw)
ax.grid(True)

# I personally prefer the last method, always set fig and axes sizes explicitely and explicitely create the axis instance   
```

# multiple plots in one figure

```python
#axes_style('light')

fig = plt.figure(figsize=(12,9))

w = .3 #width
h = .2 #height
dh = .05 #vertical spacing
dw = .05 #horizontal spacing
ncols = 3 # umber of columns

for k in range(4):
    
    row,col = divmod(k,ncols)

    ax = fig.add_axes([col*(w+dw),-row*(h+dh),w,h])
    ax.plot(df['x'],df['y'],'.-',c=c,ms=ms,lw=lw)
    ax.grid(True)
    #ax.tick_params(axis='both', which='major',labelsize=7)
```

```python
ncols = 3
nrows = 2

fig, axs = plt.subplots(ncols=ncols, nrows=nrows, figsize=(12,7))#constrained_layout=True

#
for row in range(nrows):
    for col in range(ncols):
        ax = axs[row, col]
        ax.plot(df['x'],df['y'],'.-',c=c,ms=ms,lw=lw)
        ax.grid(True)
```

```python
fig = plt.figure(figsize=(12,7))#, constrained_layout=True
spec = fig.add_gridspec(ncols=ncols, nrows=nrows,)

for row in range(nrows):
    for col in range(ncols):
        ax = fig.add_subplot(spec[row, col])
        ax.plot(df['x'],df['y'],'.-',c=c,ms=ms,lw=lw)
        ax.grid(True)
```

```python
fig = plt.figure(figsize=(12,7))#, constrained_layout=True
spec = fig.add_gridspec(ncols=ncols, nrows=nrows,)

ncols = 3 # umber of columns

for k in range(4):
    
    row,col = divmod(k,ncols)
    ax = fig.add_subplot(spec[row, col])
    ax.plot(df['x'],df['y'],'.-',c=c,ms=ms,lw=lw)
    ax.grid(True)
```

# setting axes styles

```python
plt.style.available
```

```python
fig = plt.figure(figsize=(12,12))


w=.3
h=.2
dh=.1
dw=.1
ncols=4

k=0 
for theme in plt.style.available:
    plt.style.use(theme)
    
    r,c = divmod(k,ncols)

    ax = fig.add_axes([c*(w+dw),-r*(h+dh),w,h])
    ax.plot([0,1,2,3],[1,2,5,3],'.-')
    ax.plot([0,1,2,3],[2,1,2,4],'.-')
    ax.plot([0,1,2,3],[1,3,1,2],'.-')
    
    ax.set_xlim(-.5,3.5)
    ax.set_ylim(0,6)
    ax.annotate(theme,xy=(0,1.1),xycoords='axes fraction',fontsize=12,c='.1')

    ax.grid(True)
    k+=1
    
```

## define custom style

```python
def axes_style(theme):
    
    if theme =='dark':
        fc = '.96'
        ec = '.96'
        gc = '1'
        glw = 1.5
        alw= .5
        
    if theme == 'light':
        fc = '.99'
        ec = '.5'
        gc = '.5'
        glw = .2
        alw= .5
    
    if theme == 'xlight':
        fc = '1'
        ec = '1'
        gc = '.5'
        glw = .2
        alw= 5
        
    plt.rcParams['figure.facecolor'] = '1'
        
    plt.rcParams['axes.facecolor'] = fc
    plt.rcParams['axes.edgecolor'] = ec
    plt.rcParams['axes.linewidth'] = alw

    plt.rcParams['axes.labelcolor'] = '.2'
    plt.rcParams['grid.color'] = gc
    plt.rcParams['grid.linewidth'] = glw

    plt.rcParams['xtick.major.size'] = 0
    plt.rcParams['ytick.major.size'] = 0
    
    plt.rcParams['xtick.color'] = '.2'
    plt.rcParams['ytick.color'] = '.2' 
    
    plt.rcParams['grid.linestyle'] = '-' 
```

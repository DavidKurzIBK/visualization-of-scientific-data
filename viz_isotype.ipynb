```python
fig = plt.figure(figsize=(5,5))
ax=fig.add_axes([0,0,1,1])

ncols=30 #number of columns

dw=3
dh=3
for k in range(900):
    if k>90:
        fc=(.1,.5,.9)
    else:
        fc=(.95,.3,.5)
    r,c = divmod(k,ncols)
    x0=c*(2+dw)
    y0=r*(2+dh)
    ax.plot(x0,y0,'s',ms=7,color=fc)
    
    
#ax.set_xlim(-5,150)
#ax.set_ylim(-5,150)

ax.axis('off')
```

```python
fig = plt.figure(figsize=(10,10))
ax=fig.add_axes([0,0,.2,.2])
circ = plt.Circle((1, 4.3), 1.4, color=(.4,.6,.9), alpha=1)
rect1 = plt.Rectangle((0,0),2,2,linewidth=0,edgecolor='1',facecolor=(.4,.6,.9))
rect2 = plt.Rectangle((-.5,1),3,2,linewidth=0,edgecolor='1',facecolor=(.4,.6,.9))


ax.add_patch(circ)
ax.add_patch(rect1)
ax.add_patch(rect2)


ax.set_xlim(-10,10)
ax.set_ylim(-10,10)
```

```python
def plot_glyph(ax,x0,y0,fc):
    '''
    prototype glyph
    '''
    circ = plt.Circle((x0+1, y0+4.3), 1.4, facecolor=fc)
    rect1 = plt.Rectangle((x0,y0),2,2,linewidth=0,edgecolor='1',facecolor=fc)
    rect2 = plt.Rectangle((x0-.5,y0+1),3,2,linewidth=0,facecolor=fc)

    ax.add_patch(circ)
    ax.add_patch(rect1)
    ax.add_patch(rect2)
```

```python
fig = plt.figure(figsize=(5,5))
ax=fig.add_axes([0,0,1,1])

ncols=10

dw=3
dh=5
for k in range(100):
    if k>25:
        fc=(.3,.6,.9)
    else:
        fc=(.95,.7,.3)
    r,c = divmod(k,ncols)
    x0=c*(2+dw)
    y0=r*(2+dh)
    plot_glyph(ax,x0,y0,fc)
    
    
ax.set_xlim(-5,70)
ax.set_ylim(-5,70)

ax.axis('off')
```

```python
fig = plt.figure(figsize=(12,12))
ax=fig.add_axes([0,0,1,1])

ncols=50

dw=3
dh=5
for k in range(1000):
    if k>250:
        fc=(.3,.6,.9)
    else:
        fc=(.95,.7,.3)
    r,c = divmod(k,ncols)
    x0=c*(2+dw)
    y0=r*(2+dh)
    plot_glyph(ax,x0,y0,fc)
    
    
ax.set_xlim(-5,250)
ax.set_ylim(-5,250)

ax.axis('off')
```

# axes styling

```python
fig = plt.figure(figsize=(7,7))
#plt.rcParams['figure.facecolor'] = '1'
plt.rcParams['axes.facecolor'] = '1'


plt.rcParams['axes.spines.top'] = True
plt.rcParams['axes.spines.bottom'] = True
plt.rcParams['axes.spines.left'] = True
plt.rcParams['axes.spines.right'] = True


plt.rcParams['axes.linewidth'] = 5
plt.rcParams['axes.edgecolor'] = '1'

ax = fig.add_axes([0,0,.5,.3])    
ax.grid(True)
```

```python
def axes_style(theme):
    
    if theme =='dark':
        fc = '.96'
        ec = '.96'
        gc = '1'
        glw = 1.5
        alw= .5
        
    if theme == 'light':
        fc = '.99'
        ec = '.5'
        gc = '.5'
        glw = .2
        alw= .5
    
    if theme == 'xlight':
        fc = '1'
        ec = '1'
        gc = '.5'
        glw = .2
        alw= 5
        
    plt.rcParams['figure.facecolor'] = '1'
        
    plt.rcParams['axes.facecolor'] = fc
    plt.rcParams['axes.edgecolor'] = ec
    plt.rcParams['axes.linewidth'] = alw

    plt.rcParams['axes.labelcolor'] = '.2'
    plt.rcParams['grid.color'] = gc
    plt.rcParams['grid.linewidth'] = glw

    plt.rcParams['xtick.major.size'] = 0
    plt.rcParams['ytick.major.size'] = 0
    
    plt.rcParams['xtick.color'] = '.2'
    plt.rcParams['ytick.color'] = '.2' 
    
    plt.rcParams['grid.linestyle'] = '-' 
```

# legends and direct labelling

```python
# dummy data
t=range(18)
y=np.array([3.5,5,2,7,8,3,2,3,6,4,3,7,1,2.5,2,3,2,1])
y2=np.array([3.0,4,3,8,4.3,4.1,1,2,5,2,3.4,5,2,3,4,3.5,1,4])
```

```python
#define two colors
c1=(0.011764705882352941, 0.2627450980392157, 0.8745098039215686)
c2=(1.0, 0.2784313725490196, 0.2980392156862745)
#c2=(0.6901960784313725, 0.00392156862745098, 0.28627450980392155)
```

```python

axes_style('light')
plt.rcParams['legend.frameon'] = False

h=.35
w=.44
ms=15
lw=1
fig = plt.figure(figsize=(12,7))

ax1 = fig.add_axes([0,0,w,h])

ax1.plot(t,y,'.-',c=c1,ms=ms,lw=lw)
ax1.plot(t,y2,'.-',c=c2,ms=ms,lw=lw)

ax1.set_ylim(0,10)
#ax.legend(['x1','x2'],ncol=2,bbox_to_anchor=(1,1.25))#,markerscale=0,labelspacing=40)
plt.rcParams['legend.frameon'] = True
ax1.legend(['x1','x2'])       
ax1.grid(True)


ax2 = fig.add_axes([.5,0,w,h])

ax2.plot(t,y,'.-',c=c1,ms=ms,lw=lw)
ax2.plot(t,y2,'.-',c=c2,ms=ms,lw=lw)

ax2.set_ylim(0,10)
plt.rcParams['legend.frameon'] = False
ax2.legend(['x1','x2'],ncol=2,bbox_to_anchor=(1,1.22))#,markerscale=0,labelspacing=40)
ax2.grid(True)

ax3 = fig.add_axes([1.,0,w,h])

ax3.plot(t,y,'.-',c=c1,ms=ms,lw=lw)
ax3.plot(t,y2,'.-',c=c2,ms=ms,lw=lw)

ax3.set_ylim(0,10)
#ax3.annotate()
#ax3.annotate('a)',xy=(0,1.17),xycoords='axes fraction',fontsize=12)
ax3.annotate('x1',xy=(4.3,8),xycoords='data',fontsize=17,color=c1)
ax3.annotate('x2',xy=(2.7,8),xycoords='data',fontsize=17,color=c2,ha='right')
ax3.grid(True)

ax4 = fig.add_axes([1.5,0,w,h])
ax4.plot(t,y-.6,'.-',c=c1,ms=ms,lw=lw)
ax4.plot(t,y2+4,'.-',c=c2,ms=ms,lw=lw)

ax4.set_xlim(-.5,18.5)
ax4.set_ylim(0,14)

ax4.annotate('x1',xy=(17.5,.2),xycoords='data',fontsize=17,color=c1)
ax4.annotate('x2',xy=(17.5,8),xycoords='data',fontsize=17,color=c2)#,ha='right')
ax4.grid(True)

plt.savefig(projectpath+'legends_labelling.png',bbox_inches='tight',dpi=300)
```

```python
# make time axis as datetime objects
#t_dt = pd.date_range(start='2020-6-23',periods=len(t))

t_dt = pd.date_range(start='2020-6-22',periods=50)
y=np.ones(len(t_dt))
```

```python
plt.rcParams['xtick.minor.visible'] = False
```

```python
# dummy plot for axes labels, data invisible

axes_style('dark')

fig = plt.figure(figsize=(12,7))

ax1 = fig.add_axes([1.2,0,.4,.3])
ax1.plot(t_dt,y,'.-',c=c1,ms=0,lw=0)

ax1.xaxis.set_major_locator(mdates.DayLocator(interval=3))#bymonthday=[1,7,14,28])
ax1.xaxis.set_minor_locator(mdates.MonthLocator(interval=1))

ax1.xaxis.set_major_formatter(mdates.DateFormatter('%d'))
ax1.xaxis.set_minor_formatter(mdates.DateFormatter('%b'))

for tick in ax1.xaxis.get_minorticklabels():
    tick.set_horizontalalignment("left")
    
ax1.grid(True)
ax1.tick_params(axis='both', which='major',labelsize=7)
ax1.tick_params(axis='both', which='minor', pad=15,labelsize=9)
```

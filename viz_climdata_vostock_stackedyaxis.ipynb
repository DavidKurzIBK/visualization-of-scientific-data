```python
# define colors as dictionary for better readability later
colors = {'temp':(.1,.3,.7), 'co2':(.75,.1,.25)}
```

```python
def set_style():
    '''
    set axes style
    '''
    #grid and axes color
    gacolor='.3'

    plt.rcParams['axes.facecolor'] = '1'
    plt.rcParams['axes.linewidth'] = .5
    plt.rcParams['axes.edgecolor'] = gacolor
    plt.rcParams['axes.axisbelow'] = True
    plt.rcParams['axes.labelcolor'] = gacolor


    plt.rcParams['xtick.direction']='out'
    plt.rcParams['ytick.direction']='out'

    plt.rcParams['xtick.major.size']=1
    plt.rcParams['ytick.major.size']=0

    plt.rcParams['xtick.minor.size']=0

    plt.rcParams['xtick.color']=gacolor
    plt.rcParams['ytick.color']=gacolor

    plt.rcParams['grid.color'] =gacolor
    plt.rcParams['grid.linewidth'] = .4
    plt.rcParams['grid.linestyle'] = '-'

def create_axes():
    '''
    create axes
    '''
    figsize=(12,7)
    ax1 = plt.figure(figsize=figsize).add_axes([0,0,1,.5])
    ax2 = ax1.twinx()
    
    return ax1,ax2
    
def plot_data(axes,df):
    '''
    plot T and CO2 in double y-axes
    '''
    ax1,ax2 = axes
    
    lw = 2
    ax1.plot(df['Age of ice (yr BP)'],df['Temperature variation (deg C)'],'-',c=colors['temp'],lw=lw)
    ax2.plot(df['age (yrs bp)'],df['co2'],'-',c=colors['co2'],lw=lw)

    
def format_axes(axes):
    '''
    format axes
    '''
    ax1,ax2 = axes
    
    ax1.set_yticks(np.arange(-8,4,2))
    ax1.set_ylim(-20,4)

    ax2.set_yticks(np.arange(200,350,50))
    ax2.set_ylim(150,500)

    ax1.grid(True)
    ax2.grid(True)

    # ylabels
    ax1.annotate('$\Delta$ Temp. ($^\circ$C)',xy=(-.03,.62),xycoords='axes fraction',color=colors['temp'],ha='right',rotation=90)
    ax2.annotate('CO2 (ppm)',xy=(1.04,.2),xycoords='axes fraction',color=colors['co2'],ha='left',rotation=90)

    ax1.tick_params(axis='y', colors=colors['temp'])
    ax2.tick_params(axis='y', colors=colors['co2'])
    
    #xtick labels
    xtls = np.arange(4.5e5,-5e4,-5e4)
    ax1.set_xticks(xtls)
    ax1.set_xticklabels((xtls/1e3).astype(int))
    ax1.set_xlim(4.3e5,-2e3)
    ax1.set_xlabel('Age (k yrs)')
```

```python
set_style()
axes = create_axes()
plot_data(axes,df)
format_axes(axes)
```

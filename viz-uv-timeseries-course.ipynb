# Visualize measured and modelled UV index data for 4 sites

```python
# load the datasets

datapath = 'C:/Users/david/OneDrive - uibk.ac.at/MA/VU Elective Vizualisation of scientific data/lecture material/jupyter notebooks/data/'
#datapath = 'G:/My Drive/visualization/multivar-ts/'

#list of sites
sites = ['Son','Ger','Ibk','Wie']

site = sites[0]
fn = datapath+'UVI_'+site+'.df' #dataframes pickled with py3.7
#fn = datapath+'UVI_'+site+'.csv'

df = pd.read_pickle(fn)
df.head()
```

```python
ax = plt.figure(figsize=(10,5)).add_axes([0,0,1,.5])
ax.plot(df['UV_measured'])
ax.plot(df['UV_model'])
```

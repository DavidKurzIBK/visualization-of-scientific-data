# plotting a time series
## encoding of position, length and area

```python
%matplotlib inline
import matplotlib.pyplot as plt
import pandas as pd
import numpy as np
```

```python
# set general figure style

# set grid/axis color
gacolor='.5'

plt.rcParams['axes.facecolor'] = '1'
plt.rcParams['axes.linewidth'] = .5
plt.rcParams['axes.edgecolor'] = gacolor
plt.rcParams['axes.axisbelow'] = True
plt.rcParams['axes.labelcolor'] = gacolor


plt.rcParams['xtick.direction'] = 'out'
plt.rcParams['ytick.direction'] = 'out'

plt.rcParams['xtick.major.size'] = 0
plt.rcParams['ytick.major.size'] = 0

plt.rcParams['xtick.color'] = gacolor
plt.rcParams['ytick.color'] = gacolor

plt.rcParams['grid.color'] = gacolor
plt.rcParams['grid.linewidth'] = .4
plt.rcParams['grid.linestyle'] = '-'

plt.rcParams['legend.frameon'] = False

#fontsize
fs=11
```

```python
def format_axes(ax):
    '''
    common axes formating
    '''
    ax.set_xticks(np.arange(0,20,2))
    ax.set_yticks(np.arange(0,12,2))
    
    ax.set_xlim(-.5,17.5)
    #ax.set_xlim(0,17)
    ax.set_ylim(0,10)
    
    ax.set_xlabel('time',fontsize=fs)#,labelpad=1.1)
    ax.set_ylabel('magnitude',fontsize=fs)#,labelpad=1.1)
    
    ax.tick_params(labelsize=12)
    
    ax.grid(True)
```

```python
# dummy data set
t = range(18)
y = np.array([3.5,5,2,7,8,3,2,3,6,4,3,7,1,2.5,2,3,2,1])
```

# individual figures

```python
figsize = (5,3)
#axes width, hight
w = 1
h = .5
```

```python
ax = plt.figure(figsize=figsize).add_axes([0,0,w,h])
ax.bar(t,y,color='.25',lw=0,align='center',width=.95,alpha=1)
format_axes(ax)
```

```python
ax = plt.figure(figsize=figsize).add_axes([0,0,w,h])
ax.fill_between(t,y,color=(.2,.2,.2),lw=0,alpha=.8)
format_axes(ax)
```

```python
ax = plt.figure(figsize=figsize).add_axes([0,0,w,h])

ax.step(t,y,color='.25',lw=4,where='mid')
format_axes(ax)
```

```python
ax = plt.figure(figsize=figsize).add_axes([0,0,w,h])

ax.plot(t,y,'.-',c=(.2,.2,.2),lw=0,ms=14)

format_axes(ax)
```

```python
ax = plt.figure(figsize=figsize).add_axes([0,0,w,h])

ax.plot(t,y,'.-',c=(.2,.2,.2),lw=1,ms=14)
format_axes(ax)
```

```python
ax = plt.figure(figsize=figsize).add_axes([0,0,w,h])

ax.plot(t,y,'.-',c=(.2,.2,.2),lw=4,ms=0)
format_axes(ax)
```

# all in one figure

```python
#define axes / check layout

w = .3
h = .2
dh = .07
dw = 0.06
ncols = 3

fig = plt.figure(figsize=[15,9])

axes=[]
for i in range(6):
    r,c = divmod(i,ncols)
   
    ax = fig.add_axes([c*(w+dw),-r*(h+dh),w,h])
    
    axes.append(ax)
```

```python
fig = plt.figure(figsize=[15,9])

axes=[]
for i in range(6):
    r,c = divmod(i,ncols)
   
    ax = fig.add_axes([c*(w+dw),-r*(h+dh),w,h])
    
    axes.append(ax)
    

ax0=axes[0]
ax0.plot(t,y,'.-',c=(.2,.2,.2),lw=0,ms=15)
format_axes(ax0)    

ax1=axes[2]
ax1.plot(t,y,'.-',c=(.2,.2,.2),lw=4,ms=0)
format_axes(ax1)

ax2=axes[1]
ax2.plot(t,y,'.-',c=(.2,.2,.2),lw=1,ms=13)
format_axes(ax2)

ax3=axes[3]
ax3.step(t,y,color='.25',lw=4,where='mid')
format_axes(ax3)

ax4=axes[4]
ax4.bar(t,y,color='.25',lw=0,align='center',width=.95,alpha=1)
format_axes(ax4)


ax5=axes[5]
ax5.fill_between(t,y,color=(.2,.2,.2),lw=0,alpha=.8)
format_axes(ax5)

for ax,l in zip(axes,['a)','b)','c)','d)','e)','f)']):
    ax.annotate(l, xy=(-.1,1.1), ha='right', va='top',
                xycoords='axes fraction',fontsize=fs+1,color='.1')

#plt.savefig('encoding_scheme_line_dot_area.png',bbox_inches='tight',dpi=300)
```
